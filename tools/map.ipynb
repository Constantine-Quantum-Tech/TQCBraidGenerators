```python
scale_to_rad = lambda s : 10**(-np.log2(np.log2(6)/s))
luminosity_dist = lambda rad, sigma : 0.5 + 0.5*np.exp(-rad/sigma)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sqrt, cos, atan, exp
from colorsys import hls_to_rgb

# Create figure and axes
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))

N = 100
sigma = 0.00000001

scale = np.tile(np.linspace(0, 1, N).reshape((N, 1)), N)
theta = np.tile(np.linspace(0, 2 * pi, N), (N, 1))

color = np.zeros((N, N, 3))

for t in range(N):
    for s in range(N):
        hue = t/N
        if s == 0:
            rad = 0
        else:
            rad = scale_to_rad(s/N)
        lum = luminosity_dist(rad, sigma)
        sat = 1
        color[s, t] = np.array(hls_to_rgb(hue, lum, sat))

# Plot the color map
ax.pcolormesh(theta, scale, color)

# Remove labels and ticks
ticks = [0, 0.125, 0.25, 0.5, 1]
ax.set_yticks(ticks, [f"{round(scale_to_rad(s), 4)}" if (s != 0 and s != 1) else s for s in ticks])
ax.set_rlabel_position(90)
ax.set_xticks([n*pi/4 for n in range(8)], ['0', '\u03C0/4', '\u03C0/2', '3\u03C0/4', '\u03C0', '5\u03C0/4', '3\u03C0/2', '7\u03C0/4'])
ax.set_xticks([0, pi/4, pi/2, 3*pi/4, pi, 5*pi/4, 6*pi/4, 7*pi/4], ['0', '\u03C0/4', '', '3\u03C0/4', '\u03C0', '5\u03C0/4', '3\u03C0/2', '7\u03C0/4'])

plt.show()
```
